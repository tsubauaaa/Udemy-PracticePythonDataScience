```python
# NFLのデータをサンプルとして使う
import webbrowser
website = 'http://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website)
```

```python
# クリップボードから読み込む
nfl_frame = pd.read_clipboard()
```

```python
# 列の名前が.columnsでわかる
nfl_frame.columns
```

```python
nfl_frame['First NFL Season']
```

```python
# オブジェクトの属性のような書き方も可能
nfl_frame.Team
```

```python
# 特定のカラムで新しいDataFrameを作る
DataFrame(nfl_frame, columns=['Team','First NFL Season','GP'])
```

```python
DataFrame(nfl_frame,columns=['Team','First NFL Season','GP','Stadium'])
```

```python
# 先頭だけを表示
nfl_frame.head()
```

```python
# 最後だけを表示
nfl_frame.tail()
```

```python
# indexを使って、行を取り出せる
nfl_frame.ix[3]
```

```python
# 列全体に値を代入
nfl_frame['Stadium']="Levi's Stadium"
```

```python
nfl_frame
```

```python
# 長さがあって入れば、列への代入が可能
nfl_frame["Stadium"] = np.arange(6)
nfl_frame
```

```python
# SeriesをDataFrameに追加する
stadiums = Series(["Levi's Stadium","AT&T Stadium"], index=[4,0])
```

```python
stadiums
```

```python
nfl_frame['Stadium']=stadiums
nfl_frame
```

```python
# 列を消すことも可能
del nfl_frame['Stadium']
nfl_frame
```
