```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100) + 2
```

```python
sns.distplot(data1)
sns.distplot(data2)
```

```python
sns.boxplot(data=[data1,data2])
```

```python
# 外れ値を含む
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# 水平に並べる
sns.boxplot(data=[data1,data2],whis=np.inf,orient='h')
```

```python
# 正規分布に従う乱数
data1 = stats.norm(0,5).rvs(100)

# γ分布に従う乱数
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,-1*stats.gamma(5).rvs(50)])
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
# ヴァイオリンプロットを描くと違いがわかる
sns.violinplot(data=[data1,data2])
```

```python
# バンド幅を細くする
sns.violinplot(data=data2,bw=0.01)
```

```python
# rugプロットと同じような機能もある
sns.violinplot(data=data1,inner="stick")
```
